```python
import math

def solve_quadratic():
    A = float(input("Enter coefficient A: "))
    B = float(input("Enter coefficient B: "))
    C = float(input("Enter coefficient C: "))
    
    if A == 0:
        print("This is not a quadratic equation.")
        return
    
    discriminant = B**2 - 4*A*C
    if discriminant > 0:
        root1 = (-B + math.sqrt(discriminant)) / (2 * A)
        root2 = (-B - math.sqrt(discriminant)) / (2 * A)
        print("The roots of the quadratic equation are:", root1, "and", root2)
    elif discriminant == 0:
        root = -B / (2 * A)
        print("The equation has one real root:", root)
    else:
        real_part = -B / (2 * A)
        imaginary_part = math.sqrt(abs(discriminant)) / (2 * A)
        print("The equation has complex roots:", real_part, "+", imaginary_part, "i and", real_part, "-", imaginary_part, "i")

def solve_cubic():
    A = float(input("Enter coefficient A: "))
    B = float(input("Enter coefficient B: "))
    C = float(input("Enter coefficient C: "))
    D = float(input("Enter coefficient D: "))
    
    if A == 0:
        print("This is not a cubic equation.")
        return
    
    # Convert to depressed cubic form using Cardano's method
    a = B / A
    b = C / A
    c = D / A
    
    q = (3 * b - a**2) / 9
    r = (9 * a * b - 27 * c - 2 * a**3) / 54
    discriminant = q**3 + r**2
    
    if discriminant > 0:
        s = math.copysign(1, r + math.sqrt(discriminant)) * abs(r + math.sqrt(discriminant))**(1/3)
        t = math.copysign(1, r - math.sqrt(discriminant)) * abs(r - math.sqrt(discriminant))**(1/3)
        root1 = s + t - a / 3
        print("One real root:", root1)
    else:
        theta = math.acos(r / math.sqrt(-q**3))
        root1 = 2 * math.sqrt(-q) * math.cos(theta / 3) - a / 3
        root2 = 2 * math.sqrt(-q) * math.cos((theta + 2 * math.pi) / 3) - a / 3
        root3 = 2 * math.sqrt(-q) * math.cos((theta + 4 * math.pi) / 3) - a / 3
        print("The real roots of the cubic equation are:", root1, root2, root3)

def solve_quartic():
    print("Solving quartic equations without libraries is complex and requires Ferrari's method.")
    print("This feature is not implemented.")

def main():
    while True:
        print("\nChoose an equation to solve:")
        print("1. Quadratic Equation (Ax² + Bx + C = 0)")
        print("2. Cubic Equation (Ax³ + Bx² + Cx + D = 0)")
        print("3. Quartic Equation (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
        print("4. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == '1':
            solve_quadratic()
        elif choice == '2':
            solve_cubic()
        elif choice == '3':
            solve_quartic()
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

if __name__ == "__main__":
    main()
```
